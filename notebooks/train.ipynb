### Utility Functions

```python
def get_timestamp():
    timezone = ZoneInfo("Asia/Bangkok")

    return datetime.now(timezone).strftime("%Y-%m-%d_%H:%M")
```

```python
def show_image(image, title = "Image", figsize = (6, 4)):
    if image is None:
        print("Error: Cannot load image. Please check the file path.")
    else:
        print(f"Image loaded successfully. Shape: {image.shape}")

    plt.figure(figsize = figsize)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis("off")
    plt.show()
```

### Path Variables

```python
project_dir = Path("/content/drive/MyDrive/dl-final-project")
zip_path = project_dir / "dataset.zip"
dataset_dir = Path("/content/dataset")
```

### Prepare Dataset

```python
def get_total_count(split_dir):
    total_count = len(list(split_dir.iterdir()))

    return total_count
```

```python
os.makedirs(dataset_dir, exist_ok = True)

with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(dataset_dir)

print("âœ… Extracted to:", dataset_dir)
```

```python
train_dir = dataset_dir / "images" / "train"
val_dir = dataset_dir / "images" / "val"
test_dir = dataset_dir / "images" / "test"

train_count = get_total_count(train_dir)
val_count = get_total_count(val_dir)
test_count = get_total_count(test_dir)
total_count = train_count + val_count + test_count

print("Train:", train_count)
print("Val:", val_count)
print("Test:", test_count)
print("Total:", total_count)
```

### Training

```python
training_params = {
    "project": project_dir / "runs",
    "name": get_timestamp(),
    "data": "/content/data.yaml",
    "epochs": 1000,
    "patience": 5,
    "batch": -1,
    "imgsz": 512,
    "cos_lr": True,
    "save": True,
    "save_period": 10,
    "cache": True,
    "seed": 42
}
```

```python
augmentation_params = {
    "degrees": 5.0,
    "translate": 0.1,
    "scale": 0.25,
    "shear": 0.0,
    "perspective": 0.0,
    "flipud": 0.5,
    "fliplr": 0.5,
    "mosaic": 0.5,
    "mixup": 0.1,
    "cutmix": 0.0
}
```

```python
params = {**training_params, **augmentation_params}
```
